```python
import cv2
from ultralytics import YOLO # pip install ultralytics for using YOLO model 
import numpy as np

cap = cv2.VideoCapture("Car_crash.mov") 
model = YOLO("yolov8n.pt") #yolo coco model small but very fast

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame, device="mps") # for windows using graphics card make device="0"
    result = results[0]
    print(result)
    bboxs = np.array(result.boxes.xyxy.cpu(), dtype="int")
    classes = np.array(result.boxes.cls.cpu(), dtype="int")

    for cls, bbox in zip(classes, bboxs):
        (x, y, x2, y2) = bbox 
        
        cv2.rectangle(frame, (x, y), (x2, y2), (0, 0, 255), 2) # drawing bounding box around the detected object
        cv2.putText(frame, result.names[int(result.boxes.cls[0])], (x2,y2-2), cv2.FONT_HERSHEY_PLAIN, 3, (0, 0, 255), 3) #writing text on screen
        #(x2,y2-2) gives the dimension where to write the text

    cv2.imshow("Img", frame)
    key = cv2.waitKey(1)
    if key == 27: #esc key to close
        cap.release()
        cv2.destroyAllWindows()
        break
cap.release()
cv2.destroyAllWindows()
```
